```python
B_INST, E_INST = "[INST]", "[/INST]"
B_RES, E_RES = "<function_response>", "</function_response>"
```

```python
def get_sub_comp(comp_symbol):
  company = Vnstock().stock(symbol=comp_symbol, source='TCBS').company
  list_of_sub = list(company.subsidiaries()['sub_company_name'])
  x = ",".join(list_of_sub)
  return f""" {B_INST} {B_RES}  {{\"sub_comp\": '{x}'}} {E_RES} {E_INST}"""

def get_price_stock(symbol, time):
  stock = Vnstock().stock(symbol= symbol, source='VCI')
  df = stock.quote.history(start='2024-01-01', end='2024-06-21', interval='1D')
  row_of_price = df[df['time']==time].iloc[:,1:5]
  list_of_price = row_of_price.values.tolist()[0]
  label = ['mở phiên', "cao điểm", "đáy", "đóng phiên"]
  dictionary = dict(zip(label, list_of_price))
  return f""" {B_INST} {B_RES}  {{\"price\": {dictionary} {E_RES} {E_INST}}}"""

def list_of_securites(symbol):
  stock = Vnstock().stock(symbol= symbol, source='VCI')
  df = stock.listing.all_symbols()
  return df

def list_of_stock_according_to_stock_exchange(symbol):
  stock = Vnstock().stock(symbol= symbol, source='VCI')
  df = stock.listing.symbols_by_exchange()
  return df

def list_of_stock_according_to_industry(symbol):
  stock = Vnstock().stock(symbol= symbol, source='VCI')
  df = stock.listing.symbols_by_industry()
  return df

def list_of_stock_according_to_group(symbol, group_code):
  stock = Vnstock().stock(symbol= symbol, source='VCI')
  df = stock.listing.symbols_by_group(group_code)
  return df

def list_of_industries_icb(symbol):
  stock = Vnstock().stock(symbol= symbol, source='VCI')
  df = stock.listing.industries_icb()
  return df

def intraday_quote(symbol):
  stock = Vnstock().stock(symbol= symbol, source='VCI')
  df = stock.quote.intraday(symbol=symbol, show_log=False)
  return df

def price_step_and_volume(symbol):
  stock = Vnstock().stock(symbol= symbol, source='VCI')
  df = stock.quote.price_depth(symbol=symbol)
  return df

def price_board(symbol, symbol_list):
  stock = Vnstock().stock(symbol= symbol, source='VCI')
  df = stock.trading.price_board(symbol_list)
  return df

def company_overview(symbol):
  company = Vnstock().stock(symbol= symbol, source='TSCB').company
  df = company.overview()
  return df

def company_profile(symbol):
  company = Vnstock().stock(symbol= symbol, source='TSCB').company
  df = company.profile()
  return df

def company_shareholders(symbol):
  company = Vnstock().stock(symbol= symbol, source='TSCB').company
  df = company.shareholders()
  return df

def insider_deals(symbol):
  company = Vnstock().stock(symbol= symbol, source='TSCB').company
  df = company.insider_deals()
  return df

def events(symbol):
  company = Vnstock().stock(symbol= symbol, source='TSCB').company
  df = company.events()
  return df

def news(symbol):
  company = Vnstock().stock(symbol= symbol, source='TSCB').company
  df = company.news()
  return df

def dividents(symbol):
  company = Vnstock().stock(symbol=symbol, source='TSCB').company
  df = company.dividends()
  return df

def balance_sheet(symbol, period,yearReport):
  stock  = Vnstock().stock(symbol=symbol, source='VCI')
  df = stock.finance.balance_sheet(period=period, lang='vi')
  df = df[df['Năm']==yearReport]
  return df

def income_statement(symbol, period, yearReport):
  stock  = Vnstock().stock(symbol=symbol, source='VCI')
  df = stock.finance.income_statement(period=period, lang='vi')
  df = df[df['Năm']==yearReport]
  return df


def cash_flow(symbol, period, yearReport):
  stock  = Vnstock().stock(symbol=symbol, source='VCI')
  df = stock.finance.cash_flow(period=period, lang='vi')
  df = df[df['Năm']==yearReport]
  return df


def ratio(symbol, period,yearReport):
  stock  = Vnstock().stock(symbol=symbol, source='VCI')
  df = stock.finance.ratio(period=period, lang='vi')
  df = df[df['Meta']['Năm']==yearReport]
  return df
```
